```python
import pandas as pd
import numpy as np
np.random.seed(3)
```

## 1. Line chart

```python
ts = pd.Series(np.random.randn(100), index=pd.date_range("1/1/2000", periods=100))
print(ts.head())
ts.plot();
```

```python
ts = ts.cumsum()
ts.plot()
```

### Dataframe

```python
np.random.seed(3)

df = pd.DataFrame(np.random.randn(100, 4), index=ts.index, columns=list("ABCD") )

df = df.cumsum()
print(df.head())
df.plot();
```

## 2. Bar plot

```python
df.iloc[0].plot(kind="bar");
```

Multiple bar plot

```python
df2 = pd.DataFrame(np.random.rand(10, 4), columns=["a", "b", "c", "d"])
print(df2.head())

df2.plot.bar();
```

## 5. Area plot

```python
df = pd.DataFrame(np.random.rand(10, 4), columns=["a", "b", "c", "d"])
print(df.head())

df.plot.area(stacked=True);
```

## 6. Scatter plot

```python
df = pd.DataFrame(np.random.rand(50, 4), columns=["a", "b", "c", "d"])

df["species"] = pd.Categorical(
    ["setosa"] * 20 + ["versicolor"] * 20 + ["virginica"] * 10
)

print(df.head())
print(df.tail())
df.plot.scatter(x="a", y="b");
```

```python
ax = df.plot.scatter(x="a", y="b", color="DarkBlue", label="Group 1")

df.plot.scatter(x="c", y="d", color="DarkRed", label="Group 2", ax=ax);
```

```python
df.plot.scatter(x="a", y="b", c="c", s=50);
```

###cmap https://matplotlib.org/2.0.2/examples/color/colormaps_reference.html

```python
df.plot.scatter(x="a", y="b", c="species", cmap="prism", s=df["c"] * 200);
```

```python
df.plot.scatter(x="a", y="b", s=df["c"] * 200);
```

## 7. Pie chart

```python
series = pd.Series(3 * np.random.rand(4), index=["a", "b", "c", "d"], name="series")
print(series)
series.plot.pie(figsize=(6, 6), autopct="%.2f",);
```

```python
df = pd.DataFrame(3 * np.random.rand(4, 3), index=["a", "b", "c", "d"], columns=["x", "y", "z"])
print(df.head())
df.plot.pie(subplots=True, figsize=(12, 4));
```

```python
series.plot.pie(
    labels=["AA", "BB", "CC", "DD"],
    colors=["r", "g", "b", "c"],
    autopct="%.2f",
    fontsize=12,
    figsize=(6, 6),
);
```

# Plotting tools
## 8. Scatter matrix plot

```python
from pandas.plotting import scatter_matrix

df = pd.DataFrame(np.random.randn(100, 4), columns=["a", "b", "c", "d"])

scatter_matrix(df, alpha=0.8, figsize=(8, 8), diagonal="kde");
```

### Density plot

```python
ser = pd.Series(np.random.randn(1000))

ser.plot.kde();
```
